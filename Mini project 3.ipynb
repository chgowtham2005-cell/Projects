```python
# Categorical columns
categorical_cols = [
    "cp", "fbs", "restecg",
    "exang", "slope", "ca", "thal"
]

# Numerical columns
numerical_cols = [
    "age", "treatbps", "chol",
    "thalach", "oldpeak", "sex"
]
```

# Encoding

```python
encoded_df = pd.get_dummies(df,columns=categorical_cols,drop_first=True)
```

```python
print("Original Shape:", df.shape)
print("Encoded Shape:", encoded_df.shape)
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

```python
# Convert True/False to 0/1
bool_cols = encoded_df.select_dtypes(include='bool').columns
encoded_df[bool_cols] = encoded_df[bool_cols].astype(int)

encoded_df.head()
```

# Standardized

```python
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]

scaler = StandardScaler()
encoded_df[num_cols] = scaler.fit_transform(encoded_df[num_cols])

encoded_df.head()
```

```python
encoded_df
```

# Train-test split

```python
x = encoded_df.drop(columns=['target'])
y = encoded_df.target
```

```python
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```
